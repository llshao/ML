```python
import numpy as np
import matplotlib.pyplot as plt
```

### Linear Scatter Plot

```python
# visualization: linear plot
def my_plotter(ax, x, y, line_dict = None, ax_dict = None, rc_dict = None):
    '''
    Params
    --------------------
    ax : Axes
        The axes to draw
    x : array
        The x data
    y : array
        The y data
    param_dict : dict
        Dictionary of kwargs
    Returns
    -------
    out : list
        list of artists added
    '''
    ax.plot(x, y, **line_dict)
    if ax_dict:
        ax.set(**ax_dict)
    #ax.legend(loc='best')
    #ax.legend(out, line_dict['label']);
    if rc_dict:
        plt.rcParams.update(**rc_dict)
    #ax.hold()
    #ax.hold('on')
    return ax
```

```python
line_dict = {'linestyle': '-.','marker':'^','ms':12, 
             'lw': 4, 'label': 'current'}
line1_dict = {'linestyle': ':','marker':'s','ms':12,'color':'orange', 
             'lw': 4, 'label': 'resistance'}
ax_dict = {'xlabel': 'Temperature $\mathbf{^oC}$', 
            'ylabel': 'Measured Current ($\mathbf{\mu}$A)',
            'title': 'Measured Temperature Sensor\'s Characteristics'}
ax1_dict = {'xlabel': 'Temperature $\mathbf{^oC}$', 
            'ylabel': 'Measured Resistance ($\mathbf{\Omega}$)',
            'title': 'Measured Temperature Sensor\'s Characteristics'}
rc_dict ={'legend.fontsize': 14,
          'legend.loc': 'best',
          'figure.figsize': (6, 6),
          'axes.labelsize': 14,
          'axes.labelweight': 'bold',
          'axes.grid':  'on',
          'axes.grid.which':  'both',
          'axes.titlesize': 14,
          'axes.titleweight': 'bold',
          'xtick.labelsize':14,
          'ytick.labelsize':14
}
x = np.array([27, 36, 47, 60])

y = 1e6*1/np.array([21875, 22591, 23574, 24750])
fig, ax = plt.subplots(1, 1)
ax = my_plotter(ax, x, y, 
           line_dict = line_dict, ax_dict = ax_dict, rc_dict = rc_dict)
ax.legend([line_dict['label']],loc='best', bbox_to_anchor=(0.35, 0.5))
ax1 = ax.twinx()  # instantiate a second axes that shares the same x-axis
y1 = np.array([21875, 22591, 23574, 24750])
ax = my_plotter(ax1, x, y1, 
           line_dict = line1_dict, ax_dict = ax1_dict, rc_dict = rc_dict)
ax.legend([line1_dict['label']],loc='best',bbox_to_anchor=(0.4, 0.65))
```
